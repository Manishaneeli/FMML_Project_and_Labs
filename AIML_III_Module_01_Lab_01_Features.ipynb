```python
def cleanup(text):
  text = text.lower()
  text = re.sub('[^a-z]+', '', text)
  return text

topic1_en = cleanup(topic1_en)
topic2_en = cleanup(topic2_en)
topic1_it = cleanup(topic1_it)
topic2_it = cleanup(topic2_it)
```

```python
def tuple2string(tup):
  st = ''
  for i in tup:
    st = st + i
  return st

def key2list(keys):
  #print(keys)
  return [tuple2string(i) for i in keys]

# plot the histogram
def plothistogram(ngram):
  keys = key2list(ngram.keys())
  values = list(ngram.values())
  combined = zip(keys, values)
  '''print(keys)
  print(values)
  print(combined)'''

  zipped_sorted = sorted(combined, key=lambda x: x[0])
  #print(zipped_sorted)
  keys, values = map(list, zip(*zipped_sorted))
  plt.bar(keys, values,color='yellow',edgecolor='black',alpha=0.7)

unigram_eng1 = Counter(ngrams(topic1_en,1))
plothistogram(unigram_eng1)
plt.title('English 1')
plt.show()


unigram_eng2 = Counter(ngrams(topic2_en,1))
plothistogram(unigram_eng2)
plt.title('English 2')
plt.show()

```

```python
unigram_fr1 = Counter(ngrams(topic1_it,1))
plothistogram(unigram_eng1)
plt.title('Italy 1')
plt.show()
unigram_fr2 = Counter(ngrams(topic2_it,1))
plothistogram(unigram_fr2)
plt.title('Italy 2')
plt.show()
```

```python
bigram_eng1 = Counter(ngrams(topic1_en,2)) # bigrams
plothistogram(bigram_eng1)
plt.title('English 1')
plt.show()

bigram_eng2 = Counter(ngrams(topic2_en,2))
plothistogram(bigram_eng2)
plt.title('English 2')
plt.show()

bigram_it1 = Counter(ngrams(topic1_it,2))
plothistogram(bigram_it1)
plt.title('Italy 1')
plt.show()

bigram_it2 = Counter(ngrams(topic2_it,2))
plothistogram(bigram_it2)
plt.title('Italy 2')
plt.show()
```

```python
def plotbihistogram(ngram):
  freq = np.zeros((26,26))
  for i in range(26):
    for j in range(26):
      freq[i,j] = ngram[(chr(ord('a')+i), chr(ord('a')+j))]
  plt.imshow(freq,cmap='jet')
  return freq
```

```python
bieng1 = plotbihistogram(bigram_eng1)
plt.show()
bieng2 = plotbihistogram(bigram_eng2)
```

```python
biit1 = plotbihistogram(bigram_it1)
plt.show()
biit2 = plotbihistogram(bigram_it2)
```

```python
#from IPython.core.debugger import set_trace

def ind2tup(ind):
  ind = int(ind)
  i = int(ind/26)
  j = int(ind%26)
  return (chr(ord('a')+i), chr(ord('a')+j))

def ShowTopN(bifreq, n=10):
  f = bifreq.flatten()
  arg = np.argsort(-f)
  for i in range(n):
    print(f'{ind2tup(arg[i])} : {f[arg[i]]}')
```

```python
print('\nEnglish 1:')
ShowTopN(bieng1)
print('\nEnglish 2:')
ShowTopN(bieng2)
print('\nItaly 1:')
ShowTopN(biit1)
print('\nItaly 2:')
ShowTopN(biit2)
```

```python
def cumArray(img):
  img2 = img.copy()
  for ii in range(1, img2.shape[1]):
    img2[ii,:] = img2[ii,:] + img2[ii-1,:]
  img2 = img2>0
  return img2

def getHolePixels(img):
  im1 = cumArray(img)
  im2 = np.rot90(cumArray(np.rot90(img)), 3)
  im3 = np.rot90(cumArray(np.rot90(img, 2)), 2)
  im4 = np.rot90(cumArray(np.rot90(img, 3)), 1)
  hull =  im1 & im2 & im3 & im4
  hole = hull & ~ (img>0)
  return hole
```

```python
imgs = [no1[456,:,:],no0[456,:,:]]
for img in imgs:
  plt.subplot(1,2,1)
  plt.imshow(getHolePixels(img))
  plt.subplot(1,2,2)
  plt.imshow(img)
  plt.show()
```

```python
hole1 = np.array([getHolePixels(i).sum() for i in no1])
hole0 = np.array([getHolePixels(i).sum() for i in no0])

plt.hist(hole1, alpha=0.4, color='green');
plt.hist(hole0, alpha=0.3,color='red');
```

```python
def getHullPixels(img):
  im1 = cumArray(img)
  im2 = np.rot90(cumArray(np.rot90(img)), 3)
  im3 = np.rot90(cumArray(np.rot90(img, 2)), 2)
  im4 = np.rot90(cumArray(np.rot90(img, 3)), 1)
  hull =  im1 & im2 & im3 & im4
  return hull
```

```python
imgs = [no1[456,:,:],  no0[456,:,:]]
for img in imgs:
  plt.subplot(1,2,1)
  plt.imshow(getHullPixels(img))
  plt.subplot(1,2,2)
  plt.imshow(img)
  plt.show()
```

```python
hull1 = np.array([getHullPixels(i).sum() for i in no1])
hull0 = np.array([getHullPixels(i).sum() for i in no0])

plt.hist(hull1, alpha=0.7);
plt.hist(hull0, alpha=0.7);
```

```python
def minus(a, b):
  return a & ~ b

def getBoundaryPixels(img):
  img = img.copy()>0
  rshift = np.roll(img, 1, 1)
  lshift = np.roll(img, -1 ,1)
  ushift = np.roll(img, -1, 0)
  dshift = np.roll(img, 1, 0)
  boundary = minus(img, rshift) | minus(img, lshift) | minus(img, ushift) | minus(img, dshift)
  return boundary
```

```python
imgs = [no1[456,:,:],  no0[456,:,:]]
for img in imgs:
  plt.subplot(1,2,1)
  plt.imshow(getBoundaryPixels(img))
  plt.subplot(1,2,2)
  plt.imshow(img)
  plt.show()
```

```python
bound1 = np.array([getBoundaryPixels(i).sum() for i in no1])
bound0= np.array([getBoundaryPixels(i).sum() for i in no0])

plt.hist(bound1, alpha=0.7);
plt.hist(bound0, alpha=0.7);
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA


np.random.seed(0)
n_samples = 100
feature1 = np.random.normal(loc=0, scale=1, size=n_samples)
feature2 = np.random.normal(loc=0, scale=2, size=n_samples)
feature3 = np.random.normal(loc=0, scale=0.5, size=n_samples)
feature4 = np.random.normal(loc=0, scale=3, size=n_samples)


X = np.array([feature1, feature2, feature3, feature4]).T


pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2])
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
plt.title('3D Visualization of Sample Data using PCA')
plt.show()
```

```python
#Code example for 3 rd and 4 th question
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, color, filters, measure, morphology
from skimage.transform import resize
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import random


digits = load_digits()
images = digits.images
labels = digits.target

def calculate_features(image):

    image = resize(image, (8, 8))
    image = (image > filters.threshold_otsu(image)).astype(int)  # Binarize the image


    area = np.sum(image)


    edges = filters.sobel(image)
    perimeter = measure.perimeter(edges)

    mean_intensity = np.mean(image)

    std_intensity = np.std(image)


    region_props = measure.regionprops(image)
    if region_props:
        minr, minc, maxr, maxc = region_props[0].bbox
        aspect_ratio = (maxc - minc) / (maxr - minr)
    else:
        aspect_ratio = 0

    return [
        area,
        perimeter,
        mean_intensity,
        std_intensity,
        aspect_ratio
    ]


features = np.array([calculate_features(image) for image in images])


def classify_digit(feature_vector):

    area_threshold = 1200
    perimeter_threshold = 180
    if feature_vector[0] > area_threshold and feature_vector[1] < perimeter_threshold:
        return 2
    elif feature_vector[0] < area_threshold and feature_vector[1] > perimeter_threshold:
        return 9
    else:
        return "Unknown"


random_index = random.randint(0, len(images) - 1)
random_image = images[random_index]


random_feature = calculate_features(random_image)

prediction = classify_digit(random_feature)


plt.imshow(random_image, cmap='gray')
plt.title("Classification: {}".format(prediction))
plt.show()
```
