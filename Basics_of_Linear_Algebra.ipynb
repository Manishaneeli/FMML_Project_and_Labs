#In-place Matrix Multiplication
In-place multiplication with a scalar modifies the original matrix by multiplying it with a scalar, avoiding the creation of a new matrix. This is particularly useful for memory efficiency, especially when working with large matrices.









```python
import numpy as np

matrix = np.array([[1, 2,5], [3, 4,3]])
multiplicand = 3
result = matrix * multiplicand

print(result)
```

we can use numpy,multiply function with the out argument to get In-place Multiplicatiom

```python
A = np.array([[1, 2], [3, 4]])
scalar = 2
np.multiply(A, scalar, out=A) # In-place multiplication
print(A)
```

#Comparing multi_dot method and dot method for multiplying more than 2 matrices

```python


# Matrices with random integers: entries ranging from -10 to 10
a = np.random.randint(-10, 10, (300, 4))
b = np.random.randint(-10, 10, (4, 400))
c = np.random.randint(-10, 10, (400, 6))
d = np.random.randint(-10, 10, (6, 500))
e = np.random.randint(-10, 10, (500, 150))

# Time the execution for multi_dot
time_multi_dot = %timeit -o np.linalg.multi_dot([a, b, c, d, e])

# Time the execution for dot product
time_dot_product = %timeit -o a.dot(b).dot(c).dot(d).dot(e)

# Check which multiplication takes more time and print its result
if time_multi_dot.average > time_dot_product.average:
    print("Result of the dot product multiplication (takes more time):\n", a.dot(b).dot(c).dot(d).dot(e))
else:
    print("Result of the multi_dot multiplication (takes more time):\n", np.linalg.multi_dot([a, b, c, d, e]))



```

#Matrix Power

```python
a = np.random.randint(10,20,[3,3])
# positive powers of matrix
a_4 = np.linalg.matrix_power(a, 4)
a_2 = np.linalg.matrix_power(a, 2)
# inverse of matrix
a_inv_2 = np.linalg.matrix_power(a, -3)
a_inv_1 = np.linalg.matrix_power(a, -1)
print('matrix \n', a)
print('\n matrix to the power 4\n', a_4)
print('\n matrix to the power 2\n', a_2)
print('\n matrix square inverse \n', a_inv_2)
print('\n matrix inverse \n', a_inv_1)
```



#Solving the System of Linear Equations

```python

# Coefficient Matrix
a = np.random.randint(1, 20, [3, 3])
# Dependent variable vector
b = np.array([7, 15, 20])
# solution
x = np.linalg.solve(a, b)
print('Coefficient Matrix')
print(a)
print('\nDependent Variable vector')
print(b)
print('\nSolution')
print(x)

# Check for correctness
B = a.dot(x)
print("\nCheck for Correctness:")
print("B:", B)
print("Is B equal to b?", np.allclose(B, b))
```

#SVD (Singular Value Decomposition)
Singular Value Decomposition (SVD) is a powerful tool used in machine learning to simplify data. It helps find the most important parts of the data and organizes them by importance. This makes it easier to understand and work with large amounts of information.


Think of it like finding the main themes in a story – SVD helps identify these key themes and arranges them in order of importance. By focusing on the most important themes, we can get a clearer picture of what's going on in the data.


```python
from skimage import data

image = data.cat()
print("Image properties:", image.dtype, image.min(), image.max(), image.shape)

# Display the image
plt.imshow(image)  # Use 'gray' colormap for grayscale images
plt.axis('off')  # Turn off axis
plt.show()
```

```python
# Normalize the image by dividing it by the maximum value, 255
img = image / 231.0

# Reorder the axes to have colour channels as the first dimension
img = np.transpose(img, axes=(2, 0, 1))
```

```python
U,S,V = np.linalg.svd(img)
print(U.shape, S.shape, V.shape)
```

```python
Sigma = np.zeros((3,300,451))
for i in range(3):
  np.fill_diagonal(Sigma[i,:,:], S[i,:])
print(Sigma.shape)
```

```python
# Reconstruct the original image without any dimension reduction.
reconst = U @ Sigma @ V
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```

```python
# NumPy SVD reconstruction

# Reconstruct the data by reducing the common dimensions from to 33

k = 33
reconst = U @ Sigma[:,:,:k] @ V[:,:k,:]
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```

```python


# Reconstruct the data by reducing the common dimensions to 7

k = 7
reconst = U @ Sigma[:,:,:k] @ V[:,:k,:]
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```
