##3. Can we see a clear seperation in terms of the presence/absence of disease in the features obtained from pca and tsne plots?    

PCA Plot:

The PCA plot might show some clusters of data points, but it doesn't always provide a clear separation between the two classes (presence or absence of disease). Since PCA is a linear method, it might not capture complex patterns in the data. This makes it harder to distinguish between classes if the separation isn't linear.


t-SNE Plot:

The t-SNE plot usually does a better job at showing clusters, especially in high-dimensional data. It's a non-linear method, which means it can often show a more distinct separation between classes compared to PCA. If we see clear clusters for "Disease" and "No_disease" labels in the t-SNE plot, it indicates a noticeable separation between the two classes.












##4. What is the optimal number of principal components in our case?    

To figure out the best number of principal components, we need to see how much variance each component explains. We do this by creating a scree plot or checking the cumulative variance explained by the principal components. This helps us decide how many components capture most of the important information in our data.



```python

explained_variance = pca.explained_variance_ratio_
cumulative_variance = explained_variance.cumsum()

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance) + 1), cumulative_variance, marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Scree Plot')
plt.axhline(y=0.92, color='r', linestyle='--')
plt.axhline(y=0.96, color='g', linestyle='--')
plt.grid(True)
plt.show()
```

From the scree plot, we can determine the number of components needed to explain a significant amount of the variance (e.g., 92% or 96%). Typically, we choose the number of components where the curve starts to flatten, indicating that additional components contribute less to explaining the variance.

```python
## one hot encoding of categorical features in data
def onehot_encode(df, columns, prefixes):
    df = df.copy()
    for column, prefix in zip(columns, prefixes):
        dummies = pd.get_dummies(df[column], prefix=prefix)
        df = pd.concat([df, dummies], axis=1)
        df = df.drop(column, axis=1)
    return df
```
